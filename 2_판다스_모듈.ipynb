* concat : row나 column 기준으로 단순하게 이어 붙임
* merge : 특정 고유한 키(unique id)값을 기준으로 병합

```python
df = pd.read_csv('https://bit.ly/ds-korean-idol')
df2 = pd.read_csv('https://bit.ly/ds-korean-idol-2')
```

```python
df2
```

```python
df_right = df2.drop([1, 3, 5, 7, 9], axis=0)  # 1,3,5,7,9행 (지드래곤, 뷔, 정국, 소연, 하성운)을 제거한 후
```

```python
df_right = df_right.reset_index(drop=True)  # 인덱스 정리
```

```python
df_right
```

```python
df
```

```python
pd.concat([df, df_right], axis=1)
```

```python
# merge() 사용
pd.merge(df, df_right, on='이름', how='left')  # 합칠때는 유니크한 걸로 합침  # 왼쪽데이터프레임을 중심으로 

# on=' ' : 합칠때 어떤 거로 합칠거냐
```

```python
pd.merge(df, df_right, on='이름', how='right')   
```

```python
pd.merge(df, df_right, on='이름', how='inner')

# inner 방식은 두 DataFrame에 모두 키 값이 존재하는 경우만 병합

# left와 right와 inner 세개의 차이를 알자
```

```python
pd.merge(df, df_right, on='이름', how='outer')

# outer : 왼쪽, 오른쪽 모두 보여주는 것
```

```python
# right의 컬럼명 바꾸기

df_right.columns = ['성함', '연봉', '가족수']
```

```python
df_right
```

```python
# pd.merge(df, df_right, on='이름', how='outer')

# 에러 발생
# 키의 이름 같아야 함 (왼쪽은 이름, 오른쪽은 성함으로 되어있어서 못 합침)
```

```python
# 왼쪽과 오른쪽이 다른데도 합치는 방법은 왼쪽은 left_on, 오른쪽은 right_on 붙여서 쓰면 됨

pd.merge(df, df_right, left_on='이름', right_on='성함', how='outer')
```

# **25. 원 핫 인코딩(One-Hot-Encoding)**

* 원 핫 인코딩은 한개의 요소는 True, 나머지 요소는 False로 만들어주는 기법

```python
blood_map = {'A':0, 'B':1, 'AB':2, 'O':3}
```

```python
df
```

```python
df['혈액형_code'] = df['혈액형'].map(blood_map)
```

```python
df.head()
```

```python
df['혈액형_code'].value_counts()
```

* df['혈액형_code']를 머신러닝 알고리즘에 넣어 데이터를 예측하라고 지시한다면 컴퓨터는 값들간의 관계를 스스로 형성하게 됨
* 만약 B형은 1, AB형은 2라는 값을 가지고 있을 때, 컴퓨터는 'B형 + AB형 = O형'라고 잘못된 관계를 맺을 수 있게 됨
* 별도의 column을 형성해주고 1개의 column에는 True, 나머지는 False라고 넣어줌으로 'A, B, AB, O형의 관계는 독립적이다'라는 카테고리로 표현해주는 방식

```python
pd.get_dummies(df['혈액형_code'])
```

```python
# 5-1. 노드 클래스 만들기

class Node:
  def __init__(self, value):
    self.value = value  # 여기에다가 값 저장
    self.left = None    
    self.right = None
```

```python
# 5-2. 이진 탐색 트리에 데이터 넣기

# 이진 탐색 트리 조건에 부합하게 데이터를 넣어야 함!!!
# 중복 데이터를 넣지 않음!!!

class NodeMgmt:
  def __init__(self, head):
    self.head = head

  def insert(self, value):
    self.current_node = self.head                       # 일단 head 데이터를 current_node 라고 본다. 
    while True:                                         # 무조건 무한 루프를 돌림
      if value < self.current_node.value:               # 지금 넣은 value 값이 self.current_node.value 보다 작으면 왼쪽으로 보낸다.
        if self.current_node.left != None:              # self.current_node.left 가 혹시 None 이 아니면 (데이터가 있다.)
          self.current_node = self.current_node.left    # self.current_node 에 self.current_node.left를 집어넣는다.
        else:                                           # self.current_node.left 가 혹시 None 이면 (데이터가 없다.)
          self.current_node.left = Node(value)          # 데이터가 없으면 그냥 넣어주면 된다. (self.current_node.left에다가 Node 객체를 만들어서 집어넣는다)
          break                                         # 그리고 break를 걸고 빠져나온다.
      else:                                             # 지금 넣은 value 값이 self.current_node.value 보다 크면 오른쪽으로 보낸다.
        if self.current_node.right != None:             
          self.current_node = self.current_node.right
        else:
          self.current_node.right = Node(value)
          break
```

```python
head = Node(10)  # head 에다가 Node(10)을 넣는다.

BST = NodeMgmt(head)  # BST에다가 NodeNMgmt에다가 head를 넣는다
```

```python
BST.insert(4)
```

```python
# 5-3. 이진 탐색 트리의 탐색

class NodeMgmt:
  def __init__(self, head):
    self.head = head

  def insert(self, value):
    self.current_node = self.head
    while True:
      if value < self.current_node.value:
        if self.current_node.left != None:
          self.current_node = self.current_node.left
        else:
          self.current_node.left = Node(value)
          break
      else:
        if self.current_node.right != None:
          self.current_node = self.current_node.right
        else:
          self.current_node.right = Node(value)
          break

  def search(self, value):
    self.current_node = self.head
    while self.current_node:                        # self.current_node값이 있니? 있으면 무조건 돌아
      if self.current_node.value == value:          # 만약에 self.current_node.value값이 현재 내가 집어놓은 value값이 일치하니?
        return True                                 # 일치하면 True
      elif value < self.current_node.value:  
        self.current_node = self.current_node.left
      else:
        self.current_node = self.current_node.right
    return False                                      # 아무리 돌아도 current_node가 없을때 까지 돌렸는데도 못찾았으면 False를 집어넣어준다.
```

```python
head = Node(10)

BST = NodeMgmt(head)

BST.insert(2)
BST.insert(5)
BST.insert(13)
BST.insert(19)
BST.insert(11)
BST.insert(8)
```

```python
print(BST.search(8))
```

```python
print(BST.search(19))
```

```python
print(BST.search(7))
```

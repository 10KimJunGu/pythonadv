```python
import pandas as pd

df1 = pd.read_csv('pandasdata/concat_1.csv')
df2 = pd.read_csv('pandasdata/concat_2.csv')
df3 = pd.read_csv('pandasdata/concat_3.csv')
```

```python
# concat() - 위에서 아래 방향으로 연결

row_concat = pd.concat([df1, df2, df3])
row_concat
```

```python
# 데이터프레임에 시리즈 연결

new_row = pd.Series(['n1','n2','n3','n4'])
```

```python
print(pd.concat([df1, new_row]))

# 시리즈가 새로운 행으로 추가 될 것을 기대했지만 새로운 열로 삽입이 되면서
# 구조를 오히려 복잡하게 만들고 있다. NaN(누락값) 많이 보인다.

# 행이 1개라도 반드시 데이터프레임에 담아서 연결하자.
# 시리즈에는 열이름이 없다. 따라서 새로운 열로 간주하여 0이라는 이름의 열로
# 추가한다.
```

```python
new_row_df = pd.DataFrame([['n1','n2','n3','n4']], columns = ['A','B','C','D'])

print(new_row_df)
```

```python
# 방법 1 - 데이터 프레임으로 만들어서 데이터프레임끼리 연결         concat으로 연결

pd.concat([df1, new_row_df])
```

```python
# 방법 2 - 데이터 프레임으로 만들어서 데이터프레임끼리 연결         append으로 연결


df1.append(new_row_df)
```

```python
# 방법 3 - 

data_dict = {'A': 'n1', 'B':'n2', 'C':'n3', 'D':'n4'}

print(df1.append(data_dict, ignore_index=True))

# 마지망행의 인덱스가 0 이 아닌 4로 변경 ignore_index=True
```

```python
df1
```

```python
# 새로 삽입된 데이터의 기존 인덱스를 무시하고 삽입된 새로운 인덱스로 대체된다.

pd.concat([df1, new_row_df], ignore_index=True)
```

```python
#열방향으로 데이터 연결하자

col_concat = pd.concat([df1, df2, df3], axis=1)
row_concat
```

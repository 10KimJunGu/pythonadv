```python
# seaborn, SNS 특집

import seaborn as sns
anscombe = sns.load_dataset("anscombe")
print(anscombe)
print(type(anscombe))
```

```python
import matplotlib.pyplot as plt

# 데이터 셋에서 dataset 값이 I인 데이터만 추출하여 dataset_1변수에 할당
dataset_1 = anscombe[anscombe['dataset']=='I']
dataset_1
```

```python
# 기본 그리기

plt.plot(dataset_1['x'], dataset_1['y'], 'o')

#plot은 기본적으로 선으로 그래프를 그린다. 세번째 인수로 'o'를 전달하면 점으로 표현하게 된다. 
```

```python
dataset_1 = anscombe[anscombe['dataset']=='I']
dataset_2 = anscombe[anscombe['dataset']=='II']
dataset_3 = anscombe[anscombe['dataset']=='III']
dataset_4 = anscombe[anscombe['dataset']=='IV']
```

```python
fig = plt.figure()

axes1 = fig.add_subplot(2,2,1)  # 틀 만들기
axes2 = fig.add_subplot(2,2,2)
axes3 = fig.add_subplot(2,2,3)
axes4 = fig.add_subplot(2,2,4)
```

```python
tips = sns.load_dataset("tips")
print(tips.head())
print(type(tips))
print(str(tips))

# tips 지불금액, 팁, 성별, 흡연유무, 요일, 시간, 인원 수 
```

```python
# fig 기본틀 액자
fig = plt.figure()

# 기본 틀에 내부를 -> 1개짜리
axes1 = fig.add_subplot(1,1,1)
```

```python
# fig 기본틀 액자
fig = plt.figure()

# 기본 틀에 내부를 -> 1개짜리
axes1 = fig.add_subplot(1,1,1)

axes1.hist(tips['total_bill'])

plt.show()
```

```python
# fig 기본틀 액자
fig = plt.figure()

# 기본 틀에 내부를 -> 1개짜리
axes1 = fig.add_subplot(1,1,1)

axes1.hist(tips['total_bill'])
# total bill 최종 계산 가격

axes1.set_title('Total Bill')
axes1.set_xlabel('Frequency')   # x축 이름
axes1.set_ylabel('Bill')        # y축 이름

plt.show()
```

```python
scatter_plot = plt.figure()

axes1 = scatter_plot.add_subplot(1,1,1)

axes1.scatter(tips['total_bill'], tips['tip'])

axes1.set_title('Bill vs Tip')
axes1.set_xlabel('Total bill')   # x축 이름
axes1.set_ylabel('Tip')        # y축 이름


plt.show()
```

```python
# 박스 그래프 - 성별과 팁의 상관관계

boxplot = plt.figure()

axes1 = boxplot.add_subplot(1,1,1)

axes1.boxplot([tips[tips['sex'] == "Female"]['tip'],
               tips[tips['sex'] == "Male"]['tip']],
               labels=['Female','Male'])

axes1.set_title('Tips by Sex')
axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')

plt.show()
```

```python
# Seaborn 그리기

# Seaborn 으로 그림을 그리면 matplotlib보다 좀 더 화려하게 그림을 그릴 수 있다.

import seaborn as sns

tips = sns.load_dataset('tips')
```

```python
ax = plt.subplots()

ax = sns.distplot(tips['total_bill'])

ax.set_title('Total Bill Histogram with  SNS')

plt.show()
```

```python
ax = plt.subplots()

ax = sns.distplot(tips['total_bill'], kde=False)  # kde = 실선

ax.set_title('Total Bill Histogram with  SNS')

plt.show()
```

```python
ax = plt.subplots()

ax = sns.distplot(tips['total_bill'], rug=True)  # rug = 빈도수

ax.set_title('Total Bill Histogram with  SNS')

plt.show()
```

```python
ax = plt.subplots()

ax = sns.countplot('day', data=tips) 

ax.set_title('Count of Days')

plt.show()
```

```python
ax = plt.subplots()

ax = sns.countplot('day', data=tips) 

ax.set_title('Count of Days')
ax.set_xlabel('Day of the Week')
ax.set_ylabel('Frequency')

plt.show()
```

```python
ax = plt.subplots()   #(*****)

ax = sns.regplot(x='total_bill', y = 'tip', data=tips)

ax.set_title('Total Bill and Tip')
ax.set_xlabel('Total bill')
ax.set_ylabel('Tip')

plt.show()
```

```python
# ax = plt.subplots()

ax = sns.jointplot(x='total_bill', y = 'tip', data=tips)
ax.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
ax.fig.suptitle('Joint plot of Total Bill and Tip', y=1.1)

# ax.set_title('Total Bill and Tip')
# ax.set_xlabel('Total bill')
# ax.set_ylabel('Tip')

plt.show()
```

```python
ax = sns.jointplot(x='total_bill', y = 'tip', data=tips, kind = "hex")
ax.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
ax.fig.suptitle('Joint plot of Total Bill and Tip', y=1.1)
```

```python
ax = plt.subplots()

ax = sns.kdeplot(data=tips['Total_bill'], data2=tips['tip'],shade=True)


ax.set_title('Kenrnel Desity Plot and Total Bill and Tip')
ax.set_xlabel('Total bill')
ax.set_ylabel('Tip')

plt.show()
```

```python
ax = plt.subplots()

ax = sns.barplot(x='time', y='tip', data=tips)

ax.set_title('Time and Tip')
ax.set_xlabel('Time')
ax.set_ylabel('Tip')

plt.show()
```

```python
ax = plt.subplots()

ax = sns.boxplot(x='time', y='tip', data=tips)

ax.set_title('Time and Tip')
ax.set_xlabel('Time')
ax.set_ylabel('Tip')

plt.show()
```

```python
ax = plt.subplots()

ax = sns.violinplot(x='time', y='tip', data=tips)

ax.set_title('Time and Tip')
ax.set_xlabel('Time')
ax.set_ylabel('Tip')

plt.show()
```

```python
# 관계 그래프

import warnings
warnings.simplefilter(action='ignore',category=FutureWarning)

pair_grid = sns.PairGrid(tips)
pair_grid = pair_grid.map_upper(sns.regplot)
pair_grid = pair_grid.map_lower(sns.kdeplot)
pair_grid = pair_grid.map_diag(sns.distplot, rug = True)

plt.show()
```
